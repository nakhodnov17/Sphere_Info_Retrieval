```python
from collections import Counter
import codecs
import unicodedata
import io
import re
```

```python
def create_tf(words):
    counter = Counter()
    d = dict()
    for word in words:
        counter[word] += 1 
    for word in counter:
        d[word] = (float)(counter[word]) / len(words)
    return d
```
